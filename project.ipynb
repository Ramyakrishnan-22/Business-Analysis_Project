```python

import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("C:/Users/Ramyakrishnan S/Downloads/Data_set 2 - Copy.csv")

# Step 2: List of investment columns
investment_columns = [
    'Mutual_Funds', 'Equity_Market', 'Debentures',
    'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'Stock_Marktet'
]

# Step 3: Group data by gender and sum investments
gender_investments = df.groupby('gender')[investment_columns].sum()

# Step 4: Plot grouped bar chart
gender_investments.plot(kind='bar', figsize=(12,6))
plt.title('Investment Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Investment Amount')
plt.xticks(rotation=0)
plt.legend(title='Investment Avenue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()




# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical Columns:", list(numerical_cols))
# Create a statistics table
stats_df = pd.DataFrame(index=numerical_cols)

# Mean
stats_df['Mean'] = df[numerical_cols].mean()

# Median
stats_df['Median'] = df[numerical_cols].median()

# Standard Deviation
stats_df['Std_Dev'] = df[numerical_cols].std()


print(stats_df)



```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('C:/Users/Ramyakrishnan S/Downloads/Data_set 2 - Copy.csv')
# 1. Define mapping for Investment Duration (midpoints in years)
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2.0,
    '3-5 years': 4.0,
    'More than 5 years': 7.5
}
df['Duration_Years'] = df['Duration'].map(duration_mapping)

# 2. Define mapping for Expected Returns (midpoints of percentages)
expect_mapping = {
    '10%-20%': 15.0,
    '20%-30%': 25.0,
    '30%-40%': 35.0
}
df['Expect_Numerical'] = df['Expect'].map(expect_mapping)


# --- TASK 3: DESCRIPTIVE STATISTICS FOR NUMERICAL COLUMNS ---

numerical_cols = ['age', 'Mutual_Funds', 'Equity_Market', 'Debentures', 
                  'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 
                  ] # Include 'age group' to show its constant nature
stats = df[numerical_cols].agg(['mean', 'median', 'std']).T.round(2)
print("--- DESCRIPTIVE STATISTICS ---\n", stats)


# --- TASK 4: MOST PREFERRED INVESTMENT AVENUE ---

avenue_counts = df['Avenue'].value_counts()
print("\n--- INVESTMENT AVENUE FREQUENCY ---\n", avenue_counts)
most_preferred = avenue_counts.index[0]
print(f"\nMost Preferred Avenue: {most_preferred} ({avenue_counts.iloc[0]} respondents)")


# --- TASK 8: AVERAGE INVESTMENT DURATION ---

average_duration = df['Duration_Years'].mean()
print(f"\n--- AVERAGE INVESTMENT DURATION ---\nCalculated Average Duration: {average_duration:.2f} years")


# --- TASK 9: CORRELATION ANALYSIS (AGE, DURATION, EXPECTATIONS) ---

correlation_matrix = df[['age', 'Duration_Years', 'Expect_Numerical']].corr()
print("\n--- CORRELATION MATRIX (age, Duration, Expectation) ---\n", correlation_matrix.round(2))

# Visualization: Matplotlib Heatmap (compatible code)
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(correlation_matrix, cmap='coolwarm')
fig.colorbar(cax, label='Correlation Coefficient')
ticks = np.arange(len(correlation_matrix.columns))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(correlation_matrix.columns, rotation=45, ha='left')
ax.set_yticklabels(correlation_matrix.columns)

# Add correlation values (annotations)
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        ax.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}",
                ha="center", va="center", color="black")

plt.title('Correlation Heatmap (age, Duration, Expectation)', pad=20)
plt.tight_layout()
plt.savefig('final_correlation_heatmap.png')
# Note: The figure is saved, but plt.show() is omitted as per guidelines.
# The image file is available for review.
```
